```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
di["Glucose"]=di["Glucose"].replace(0,np.nan)
di["BloodPressure"]=di["BloodPressure"].replace(0,np.nan)
di["SkinThickness"]=di["SkinThickness"].replace(0,np.nan)
di["Insulin"]=di["Insulin"].replace(0,np.nan)
```

```python
di.head()
```

```python
di.isnull().sum()
```

```python
di.median()
```

```python
di.fillna(di.median(), inplace=True)
di.head()
```

```python
di.isna().sum()
```

```python
plt.figure(figsize=(20,30))

plt.subplot(4,4,1)
sns.boxplot(di["Pregnancies"])

plt.subplot(4,4,2)
sns.boxplot(di["Glucose"])

plt.subplot(4,4,3)
sns.boxplot(di["BloodPressure"])

plt.subplot(4,4,4)
sns.boxplot(di["SkinThickness"])

plt.subplot(4,4,5)
sns.boxplot(di["Insulin"])

plt.subplot(4,4,6)
sns.boxplot(di["Age"])


plt.show()
```

```python
di["Pregnancies"]=di["Pregnancies"].clip(lower=di["Pregnancies"].quantile(0.05), upper=di["Pregnancies"].quantile(0.95))
di["BloodPressure"]=di["BloodPressure"].clip(lower=di["BloodPressure"].quantile(0.05), upper=di["BloodPressure"].quantile(0.95))
di["SkinThickness"]=di["SkinThickness"].clip(lower=di["SkinThickness"].quantile(0.05), upper=di["SkinThickness"].quantile(0.95))
di["Insulin"]=di["Insulin"].clip(lower=di["Insulin"].quantile(0.05), upper=di["Insulin"].quantile(0.95))
di["BMI"]=di["BMI"].clip(lower=di["BMI"].quantile(0.05), upper=di["BMI"].quantile(0.95))
di["DiabetesPedigreeFunction"]=di["DiabetesPedigreeFunction"].clip(lower=di["DiabetesPedigreeFunction"].quantile(0.05), upper=di["DiabetesPedigreeFunction"].quantile(0.95))
di["Age"]=di["Age"].clip(lower=di["Age"].quantile(0.05), upper=di["Age"].quantile(0.95))
```

```python
plt.figure(figsize=(20,30))

plt.subplot(4,4,1)
sns.boxplot(di["Pregnancies"])

plt.subplot(4,4,2)
sns.boxplot(di["Glucose"])

plt.subplot(4,4,3)
sns.boxplot(di["BloodPressure"])

plt.subplot(4,4,4)
sns.boxplot(di["SkinThickness"])

plt.subplot(4,4,5)
sns.boxplot(di["Insulin"])

plt.subplot(4,4,6)
sns.boxplot(di["Age"])
plt.subplot(4,4,7)
sns.boxplot(di["BMI"])


plt.show()
```
